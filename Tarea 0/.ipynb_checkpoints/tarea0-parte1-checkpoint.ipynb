### Pregunta 4

Considere que se quiere mostrar la funcion 6cos(tetha/2)sin(tetha) entre 0 y 4pi con 200 puntos equiespaciados. Como podemos conseguir el objetivo de forma vectorizada y eficiente. 

Tiene el siguiente codigo base:

```python
Theta = np.linspace(0, 4*np.pi, 200)
X = 6 * np.cos( .5 * Theta ) * np.cos( Theta)
# Aca falta una linea
Y = 6 * np.cos( .5 * Theta ) * np.sin( Theta)
fig, ax = plt.subplots()
ax.plot(X, Y)
plt.grid()
plt.show()
```

# NUMPY

## Como crear arrays

```python
np.zeros(2)
np.ones(2)

# Un empty array: Contenido inicial aleatorio y depende del estado de la memoria.
np.empty(2)

# Array con rango de elementos
np.arange(4)
## np.arange(first, last, salt)
np.arange(2,9,2) 

np.linspace(0, 10, num=5)

# Data types
x = np.ones(2, dtype=np.int64)
x
```

## Como agregar, quitar y ordenar elementos

### Ordenar elemento

```python

a = np.array([2,10,3,7,4])

np.sort(a)
```

### Concatenar arrays

```python

a = np.array([1, 2, 3, 4])
b = np.array([5, 6, 7, 8])

np.concatenate((a, b))
```

### Eliminar elementos

Para eliminar elementos de un array, es sencillo usar la indexacion para seleccionar los elementos que desea conservar

## Operaciones basicas sobre arrays


### Sumar arrays

```python
data = np.array([1, 2])
ones = np.ones(2, dtype=int)
data + ones
```

### Sumar elementos de columnas o filas

```python
b = np.array([[1, 1], [2, 2]])

# Suma de filas
b.sum(axis=0)

# Suma de columnas
b.sum(axis=1)

```
